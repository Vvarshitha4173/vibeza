```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
from sklearn.cluster import KMeans

# Example dataset
df = pd.DataFrame({'Age': [25, 34, 45, 23, 35], 'Income($)': [50000, 60000, 120000, 30000, 70000]})

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(df[['Age', 'Income($)']])  # Ensure clustering is performed

print(df.head())  # Check if the 'cluster' column exists
```

```python
import pandas as pd
a="/content/income.csv"
df=pd.read_csv(a)
df
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Scatter Plot of Age vs. Income')
plt.grid(True)
plt.show()
```

```python

from sklearn.cluster import KMeans

km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
km.cluster_centers_
```

```python
df1=df[df['cluster']==0]
df2=df[df['cluster']==1]
df3=df[df['cluster']==2]
#df4=df[df['cluster']==3]
plt.scatter(df1.Age,df1['Income($)'],color='blue')
plt.scatter(df2.Age,df2['Income($)'],color='red')
plt.scatter(df3.Age,df3['Income($)'],color='pink')
#plt.scatter(df4.Age,df4['Income($)'],color='green')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color="purple",marker='*',label='centroid')
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```

```python
df1 = df[df["cluster"] == 0]
df2 = df[df["cluster"] == 1]
df3 = df[df["cluster"] == 2]

import matplotlib.pyplot as plt
plt.scatter(df1["Age"], df1["Income($)"], color="blue")
plt.scatter(df2["Age"], df2["Income($)"], color="red")
plt.scatter(df3["Age"], df3["Income($)"], color="green")

plt.xlabel("Age")
plt.ylabel("Income($)")
plt.show()
```

```python
aa=MinMaxScaler()
aa.fit(df[['Income($)']])
df['Income($)']=aa.transform(df[['Income($)']])
aa.fit(df[["Age"]])
df["Age"]=aa.transform(df[["Age"]])
```
