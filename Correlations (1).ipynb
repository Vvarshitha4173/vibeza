```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

np.random.seed(42)
data={
    "Study hours":np.random.randint(1,10,10),
    "Test Score":np.random.randint(50,100,10),
    "Age":np.random.randint(18,25,10),
    "Sleep Hours":np.random.randint(4,8,10),
    "Exercise Hours":np.random.randint(1,5,10),
    "Marks":np.random.randint(50,100,10)
}

df=pd.DataFrame(data)
df
```

```python
fe1="Study hours"
fe2="Marks"
corr=df[fe1].corr(df[fe2])
print(f"Correlation between {fe1} and {fe2} : {corr} ")
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
plt.show()
```
